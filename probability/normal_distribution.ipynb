```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

# PDF

### full code

Normal Distrubution : mu = 0 sigma = 1

* กำหนดไว้ใน function เพื่อ กำหนดค่าพื้นฐาน

```python
SQRT_TWO_PI = math.sqrt(2 * math.pi)
```

```python
def normal_pdf(x:float ,mu:float = 0 ,sigma:float = 1)->float:
    return (math.exp(-(x-mu)**2/2/sigma**2) /(SQRT_TWO_PI * sigma))
```

```python
xs = [x/10 for x in range(-50,50)]
```

```python
plt.plot(xs,[normal_pdf(x,sigma=1) for x in xs],'-',label = 'mu = 0, sigma = 1')
plt.plot(xs,[normal_pdf(x,sigma=2) for x in xs],'--',label = 'mu = 0, sigma = 2')
plt.plot(xs,[normal_pdf(x,sigma=0.5) for x in xs],':',label = 'mu = 0, sigma = 0.5')
plt.plot(xs,[normal_pdf(x,mu=-1) for x in xs],'-.',label = 'mu = -1, sigma = 1')
plt.legend()
plt.title('Various Normal pdfs')
plt.show()
```

### scipy.stat

norm.pdf(xs,loc,scale)

* loc = mu ,scale = sigma

```python
from scipy.stats import norm
```

```python
plt.plot(xs,norm.pdf(xs,0,1),'-',label = 'mu = 0 , sigma = 1')
plt.plot(xs,norm.pdf(xs,0,2),'--',label = 'mu = 0 , sigma = 2')
plt.plot(xs,norm.pdf(xs,0,0.5),':',label = 'mu = 0 , sigma = 0.5')
plt.plot(xs,norm.pdf(xs,-1,1),'-.',label = ' mu = -1 , sigma = 1')
plt.legend()
plt.title('Various Normal pdfs (scipy.stats)')
plt.show()
```

# CDF

### full code

```python
def normal_cdf(x:float,mu:float = 0 ,sigma:float = 1) -> float:
    return (1+ math.erf((x - mu )/ math.sqrt(2)/sigma))/2
```

```python
xs = [x/10 for x in range(-50,50)]
```

```python
plt.plot(xs,[normal_cdf(x,sigma=1) for x in xs],'-',label = 'mu = 0 , sigma = 1')
plt.plot(xs,[normal_cdf(x,sigma=2) for x in xs],'--',label = 'mu = 0 , sigma = 2')
plt.plot(xs,[normal_cdf(x,sigma=0.5) for x in xs],':',label = 'mu = 0 , sigma = 0.5')
plt.plot(xs,[normal_cdf(x,mu = -1) for x in xs],'-.',label = 'mu = -1 , sigma = 1')
plt.legend(loc = 4) #bottom right
plt.title('Various Normal cdfs')
plt.show()
```

### scipy.stats.norm

```python
plt.plot(xs,norm.cdf(xs,0,1),'-',label = 'mu = 0 , sigma = 1')
plt.plot(xs,norm.cdf(xs,0,2),'--',label = 'mu = 0 , sigma = 2')
plt.plot(xs,norm.cdf(xs,0,0.5),':',label = 'mu = 0 , sigma = 0.5')
plt.plot(xs,norm.cdf(xs,-1,1),'-.',label = 'mu = -1 , sigma = 1')
plt.legend(loc = 4) #bottom right
plt.title('Various Normal cdfs (scipy)')
plt.show()
```

# RVS

```python
xs = [x/10 for x in xs]
```

```python
x_rvs = norm.rvs(loc = 0,scale = 1 , size = 100000 , random_state = 3)
```

```python
fig , ax = plt.subplots(figsize = (16,6))
ax.hist(x_rvs, bins = 1000)
plt.show()
```

```python
mu , sigma = norm.fit(x_rvs)

xs = [x/10 for x in range(-50,50)]
fig , ax= plt.subplots(figsize = (16,6))
ax.hist(x_rvs , bins = 1000 ,density=True)
ax.plot(xs , norm.pdf(xs , loc = mu, scale = sigma) , lw = 5)
plt.show()
```
