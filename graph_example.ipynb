# Example 01

```python
import matplotlib.pyplot as plt
```

```python
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2,543.3,2075.9,2862.2,5979.6,10289.7,14958.3]
```

```python
plt.plot(years,gdp,'r--v')
for year,g in zip(years,gdp):
    plt.annotate((year,g),xy=(year,g),xytext = (5,10), textcoords = 'offset points',size = 7)
```
