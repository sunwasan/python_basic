# Line Graph 02

```python
variance = [1,2,4,8,16,32,64,128,256]
bias_square = [256,128,64,32,16,8,4,2,1]
total_error = [x+y for x,y in zip(variance,bias_square)]
xs = [ i for i,_ in enumerate(variance)]
```

```python
plt.plot(xs ,variance, 'g-' , label = 'variance')
plt.plot(xs , bias_square , 'r-.' , label = 'bias^2')
plt.plot(xs , total_error , 'b:' , label = 'total error')

plt.legend()
plt.xlabel('model complexity')
plt.xticks([])
plt.yticks([i for i in range (0,350,50)])
plt.title('The Bias-Variance Tradeoff')
```

# Bar Exam

```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
histrogram = Counter(min(grade // 10*10,90)for grade in grades)
```

```python
histrogram
```

```python
plt.bar([x+5 for x in histrogram.keys()], histrogram.values(),)
```

ขยายกราฟ และตีกรอบ

```python
plt.bar([x+5 for x in histrogram.keys()], histrogram.values(),10,edgecolor =(0,0,0))
```

ปรับ  y ticks ให้เหมาะสม

```python
plt.bar([x+5 for x in histrogram.keys()], histrogram.values(),10,edgecolor =(0,0,0))
plt.axis([-5,105,0,5])
```

plt.axis(xmin,xmax,ymin,ymax) เพื่อจัดแกนให้เหมาะสม

```python
plt.bar([x+5 for x in histrogram.keys()], histrogram.values(),10,edgecolor =(0,0,0))
plt.axis([-5,105,0,5])
plt.xticks([10*i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('#  of Students')
plt.title('Distribution of Exam 1 Grades')
```

# Scatter Graph

```python
friends = [70,65,72,63,71,64,60,64,67]
minutes = [175,170,205,120,220,130,105,145,190]
labels = ['a','b','c','d','e','f','g','h','i']
```

```python
plt.scatter(friends,minutes)

for label,friend_count,minute_cout in zip(labels,friends,minutes):
    plt.annotate(label,
                xy = (friend_count,minute_cout),
                xytext=(5,-5),
                textcoords = 'offset points'  )

plt.title('Daily Minutes vs. Number of Friends')
plt.xlabel('# of friend')
plt.ylabel('daily minutes spent on the site')
```

# Scatter 02

แผนภูมิซึ่งมีแกนเทียบกันไม่ได้

```python
test_1_grades = [99,90,85,97,80]
test_2_grades = [100,85,60,90,70]
labels = ['a','b','c','d','e']
```

```python
plt.scatter(test_1_grades,test_2_grades)
plt.title('Axes are Comparable')
plt.xlabel('test 1 grade')
plt.ylabel('test 2 grade')
for label,test1,test2 in zip(labels,test_1_grades,test_2_grades):
    plt.annotate(label , xy = (test1,test2) , xytext = (5,-5),textcoords = 'offset points')
```

เพิ่มคำสั่ง plt.axies('equal')

```python
plt.scatter(test_1_grades,test_2_grades)
plt.title('Axes are Comparable')
plt.xlabel('test 1 grade')
plt.ylabel('test 2 grade')
plt.axis('equal')

for label,test1,test2 in zip(labels,test_1_grades,test_2_grades):
    plt.annotate(label , xy = (test1,test2) , xytext = (5,-5),textcoords = 'offset points')
```
